```python
import pandas as pd
```

```python
mobile = pd.read_csv("mobile_price_data.csv")
```

Briefly explore the raw dataset

```python
mobile.head()
```

First of all, we remove those units in the columns, by using regex

```python
import re
mobile["disp_size"] = mobile["disp_size"].apply(lambda x: re.sub(r'(.+)(cm)(.+)',r'\1', x))
mobile["mp_speed"] = mobile["mp_speed"].apply(lambda x: re.sub(r'(.+)( GHz)',r'\1', x))
mobile["int_memory"] = mobile["int_memory"].apply(lambda x: re.sub(r'(.+)( GB)',r'\1', x))
mobile["ram"] = mobile["ram"].apply(lambda x: re.sub(r'(.+)( GB)',r'\1', x))
mobile["battery_power"] = mobile["battery_power"].apply(lambda x: re.sub(r'(.+)( mAh)',r'\1', x))
mobile["mob_width"] = mobile["mob_width"].apply(lambda x: re.sub(r'(.+)( mm)',r'\1', x))
mobile["mob_height"] = mobile["mob_height"].apply(lambda x: re.sub(r'(.+)( mm)',r'\1', x))
mobile["mob_depth"] = mobile["mob_depth"].apply(lambda x: re.sub(r'(.+)( mm)',r'\1', x))
mobile["mob_weight"] = mobile["mob_weight"].apply(lambda x: re.sub(r'(.+)( g)',r'\1', x)) 
```

Splitting "mobile_name" into two columns: "mobile_brand" and "model_name"

```python
mobile[['mobile_brand', 'model_name']] = mobile['mobile_name'].str.split(' ', 1, expand=True)
mobile.drop('mobile_name', inplace=True, axis=1)

mobile["model_name"] = mobile["model_name"].apply(lambda x: re.sub(r'(.*?) \((.*)',r'\1', x))
```

Splitting "resolution", "f_cam" and "p_cam", for the ease of data analysis

```python
mobile[['resolution_width', 'resolution_height']] = mobile['resolution'].str.split(r'x|×|\*', 1, expand=True)
mobile.drop('resolution', inplace=True, axis=1)
mobile["resolution_height"] = mobile["resolution_height"].astype(str)
mobile["resolution_height"] = mobile["resolution_height"].apply(lambda x: re.sub(r'(\d+)(.*)',r'\1', x))

mobile[['f_cam_1', 'f_cam_2']] = mobile['f_cam'].str.split('+', 1, expand=True)
mobile.drop('f_cam', inplace=True, axis=1)
mobile["f_cam_1"] = mobile["f_cam_1"].astype(str)
mobile["f_cam_2"] = mobile["f_cam_2"].astype(str)
mobile["f_cam_1"] = mobile["f_cam_1"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["f_cam_1"] = mobile["f_cam_1"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["f_cam_1"] = mobile["f_cam_1"].apply(lambda x: re.sub(r'(\D+)',r'', x))
mobile["f_cam_2"] = mobile["f_cam_2"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["f_cam_2"] = mobile["f_cam_2"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["f_cam_2"] = mobile["f_cam_2"].apply(lambda x: re.sub(r'(\D+)',r'', x))

mobile[['p_cam_1', 'p_cam_2', 'p_cam_3', 'p_cam_4']] = mobile['p_cam'].str.split('+', 4, expand=True)
mobile.drop('p_cam', inplace=True, axis=1)
mobile["p_cam_1"] = mobile["p_cam_1"].astype(str)
mobile["p_cam_2"] = mobile["p_cam_2"].astype(str)
mobile["p_cam_3"] = mobile["p_cam_3"].astype(str)
mobile["p_cam_4"] = mobile["p_cam_4"].astype(str)
mobile["p_cam_1"] = mobile["p_cam_1"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["p_cam_1"] = mobile["p_cam_1"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["p_cam_1"] = mobile["p_cam_1"].apply(lambda x: re.sub(r'(\D+)',r'', x))
mobile["p_cam_2"] = mobile["p_cam_2"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["p_cam_2"] = mobile["p_cam_2"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["p_cam_2"] = mobile["p_cam_2"].apply(lambda x: re.sub(r'(\D+)',r'', x))
mobile["p_cam_3"] = mobile["p_cam_3"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["p_cam_3"] = mobile["p_cam_3"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["p_cam_3"] = mobile["p_cam_3"].apply(lambda x: re.sub(r'(\D+)',r'', x))
mobile["p_cam_4"] = mobile["p_cam_4"].apply(lambda x: re.sub(r'(\D+)(\d+)(\D+)',r'\2', x))
mobile["p_cam_4"] = mobile["p_cam_4"].apply(lambda x: re.sub(r'(\d+)(\D+)',r'\1', x))
mobile["p_cam_4"] = mobile["p_cam_4"].apply(lambda x: re.sub(r'(\D+)',r'', x))
```

For the "network" column, we just keep the newer and better network type, i.e.: 5G>4G VOLTE>4G>3G>2G

```python
list1 = [] 
for x in mobile["network"]:
    if "5G" in x:
        list1.append("5G")
    elif "4G VOLTE" in x:
        list1.append("4G VOLTE")
    elif "4G" in x:
        list1.append("4G")
    elif "3G" in x:
        list1.append("3G")
    elif "2G" in x:
        list1.append("2G")
mobile["network"] = list1
mobile.rename(columns={'network': 'network_top'}, inplace=True)
```

Briefly explore our cleaned dataset, to check whether there is any error or mistake during our cleaning process

Remove some extreme data

```python
mobile = mobile.loc[mobile["mobile_brand"] != "MI3"]
mobile = mobile.loc[mobile["p_cam_1"] != "4812" ]
```

Last but not least, we convert the "mobile_price", from Indian Rupee into Singapore Dollar. Using the currency rate on 11 Mar 2022.

```python
mobile["mobile_price"] = mobile["mobile_price"].apply(lambda x: re.sub(r'(.*)\,(.*)',r'\1\2', x))
mobile["mobile_price"] = mobile["mobile_price"].apply(lambda x: re.sub(r'.(\d+)',r'\1', x))
mobile["mobile_price"] = pd.to_numeric(mobile["mobile_price"])
mobile["mobile_price"] = mobile["mobile_price"].apply(lambda x: round(x*0.0178), x)
```
